```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def stability_function(z):
    """Calculates the stability function R(z) for the classical RK method."""
    return 1 + z + z**2 / 2 + z**3 / 6 + z**4 / 24


# Create a grid of complex numbers
x = np.linspace(-3, 1, 1000)
y = np.linspace(-3, 3, 1000)
X, Y = np.meshgrid(x, y)
Z = X + 1j * Y

# Calculate the stability function at each point
R = stability_function(Z)

# Plot and fill the contour |R(z)| < 1
plt.contourf(X, Y, np.abs(R), levels=[0, 1], colors="blue")
plt.xlabel("Re(z)")
plt.ylabel("Im(z)")
plt.savefig("2_stability_domain.png", dpi=300)
```
